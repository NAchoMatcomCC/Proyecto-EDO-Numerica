## 1. Puntos de Equilibrio en función de μ

**Ecuación:**  
$$
\frac{dz}{dt} = \mu - z^2
$$

**Condición de equilibrio:**  
$$
\frac{dz}{dt} = 0 \quad \Rightarrow \quad \mu - z^2 = 0
$$

**Puntos de equilibrio:**  
$$
z^2 = \mu
$$

- **Si ($\mu > 0$):**  
  Dos puntos de equilibrio:  
  $$
  z = \sqrt{\mu} \quad \text{y} \quad z = -\sqrt{\mu}
  $$

- **Si ($\mu = 0$):**  
  Un punto de equilibrio:  
  $$
  z = 0
  $$

- **Si ($\mu < 0$):**  
  No hay puntos de equilibrio reales.

---

## 2. Clasificación de Estabilidad

**Derivada para análisis de estabilidad:**
$$
\frac{d}{dz}(\mu - z^2) = -2z
$$

**Evaluación en puntos de equilibrio:**

**Para $\mu > 0$:**
- En $z = \sqrt{\mu}$: $-2\sqrt{\mu} < 0$ → **PUNTO ESTABLE**
- En $z = -\sqrt{\mu}$: $2\sqrt{\mu} > 0$ → **PUNTO INESTABLE**

**Para $\mu = 0$:**
- En $z = 0$: $0$ → **CASO CRÍTICO**

---

## 3. Diagrama de Bifurcación

**Ejes:**
- Horizontal: parámetro $\mu$
- Vertical: variable $z$

**Curvas de equilibrio:**
- $z = \sqrt{\mu}$ → línea CONTINUA (estable)
- $z = -\sqrt{\mu}$ → línea DISCONTINUA (inestable)

**Punto de bifurcación:**
- Ubicación: $\mu = 0$
- Tipo: **BIFURCACIÓN NODO-SILLA**


```python
import numpy as np
import matplotlib.pyplot as plt

mu = np.linspace(-1, 2, 500)
z_stable = np.sqrt(mu[mu >= 0])
z_unstable = -np.sqrt(mu[mu >= 0])

plt.plot(mu[mu >= 0], z_stable, 'b-', label='Estable')
plt.plot(mu[mu >= 0], z_unstable, 'r--', label='Inestable')
plt.axvline(x=0, color='k', linestyle=':', alpha=0.5)
plt.xlabel('μ')
plt.ylabel('z')
plt.legend()
plt.grid(True)
plt.show()
```

## 4. Interpretación Física

**Contexto:** Movimiento con resistencia no lineal

- **$\mu > 0$:** Sistema alcanza VELOCIDAD TERMINAL ESTABLE ($z = \sqrt{\mu}$)
- **$\mu = 0$:** Transición crítica entre regímenes dinámicos
- **$\mu < 0$:** NO EXISTE velocidad terminal - comportamiento no acotado

**Significado:** La bifurcación en $\mu = 0$ marca el umbral donde desaparece la posibilidad de equilibrio dinámico.
